# Read original files and extract gene_id, copy_number, max_copy_number, labels(lung cancer type). Save data to .pkl file.

```python
patient_ids = []
gene_ids = []
labels = []     # LUSC: 0, LUAD: 1
copy_number_matrix = None
max_copy_number_matrix = None
count = 0
for path in pathlib.Path(dir).iterdir():
    if os.path.isdir(path):
        for file in pathlib.Path(path).iterdir():
            if file.name.endswith('.tsv'):
                count += 1
                lst = file.name.split('.')
                if 'TCGA' in lst[0]:
                    patient_id = lst[1]
                    patient_ids.append(patient_id)
                    if 'LUSC' in lst[0]:
                        labels.append(0) 
                    else:
                        labels.append(1) 

                if 'ASCAT' in lst[2]:
                    break

                df = pd.read_csv(file, sep='\t')
                if not gene_ids:
                  gene_ids = df['gene_id'].tolist()
                # extract copy_number
                copy_number_lst = df['copy_number'].tolist()
                ary = np.array([copy_number_lst])
                copy_number_matrix = ary if copy_number_matrix is None else np.append(copy_number_matrix, ary, axis=0)
                # extract max_copy_number
                max_copy_number_lst = df['max_copy_number'].tolist()
                ary = np.array([max_copy_number_lst])
                max_copy_number_matrix = ary if max_copy_number_matrix is None else np.append(max_copy_number_matrix, ary, axis=0)
                print(count, '-', 'patient_id: ', patient_id, ' : ', df.shape, copy_number_matrix.shape)

print(len(patient_ids), len(gene_ids), len(labels), copy_number_matrix.shape)

# write data to pkl file
data = {'copy_number_matrix': copy_number_matrix,
        'patient_ids': patient_ids,
        'labels': labels,
        'gene_ids': gene_ids}

with open('/drive/My Drive/CS286/cs286_project_lung cancer/data/processed_data.pkl', 'wb') as f:
    pickle.dump(data, f)
    print('data(copy_number) saved!')

data1 = {'max_copy_number_matrix': max_copy_number_matrix,
        'patient_ids': patient_ids,
        'labels': labels,
        'gene_ids': gene_ids}

with open('/drive/My Drive/CS286/cs286_project_lung cancer/data/processed_data(max_copy_number).pkl', 'wb') as f:
    pickle.dump(data1, f)
    print('data(max_copy_number) saved!')
```
